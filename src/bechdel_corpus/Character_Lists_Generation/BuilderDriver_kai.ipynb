```python
# print(names_set)
names_dict = dict()
tmp_set = names_set.copy()

def add_alias(alias, name, names_dict, tmp_set):
    if name not in names_dict:
        names_dict[name] = list()
    names_dict[name].append(alias)
    tmp_set.remove(alias)

# simple mentions
# all names from names_set
lst = [
 'Mrs. Smith', 'Admiral Baldwin', "Anne Elliot", 'James Benwick'
]

for n in lst:
    formatted_n = n.replace("’", "").replace("“","").strip()
    add_alias(n, formatted_n, names_dict, tmp_set)
    
complex_set = tmp_set.copy()

# aliases for names in names_set
for name in complex_set:
        
    if 'Benwick' in name:
        add_alias(name, 'James Benwick', names_dict, tmp_set)
    elif 'Brigden' in name:
        add_alias(name, 'Captain Brigden', names_dict, tmp_set)
    elif 'Croft' in name:
        add_alias(name, 'Captain Croft', names_dict, tmp_set)
    elif 'Wentworth' in name:
        add_alias(name, 'Captain Frederick Wentworth', names_dict, tmp_set)
    elif 'Harville' in name:
        add_alias(name, 'Captain Harville', names_dict, tmp_set)
    elif 'Hayter' in name:
        add_alias(name, 'Charles Hayter', names_dict, tmp_set)
    elif 'Henrietta' in name:
        add_alias(name, 'Henrietta Musgrove', names_dict, tmp_set)
    elif 'Smith' in name:
        add_alias(name, 'Charles Smith', names_dict, tmp_set)
    elif 'Clay' in name:
        add_alias(name, 'Penelope Clay', names_dict, tmp_set)
    elif 'Wallis' in name:
        add_alias(name, 'Colonel Wallis', names_dict, tmp_set)
    elif 'Anne' in name:
        add_alias(name, 'Anne Elliot', names_dict, tmp_set)
    elif 'Louisa' in name:
        add_alias(name, 'Louisa Musgrove', names_dict, tmp_set)
    elif 'Dick' in name:
        add_alias(name, 'Dick Musgrove', names_dict, tmp_set)
    elif 'Elizabeth' in name:
        add_alias(name, 'Elizabeth Elliot', names_dict, tmp_set)
    elif 'Shirley' in name:
        add_alias(name, 'Rev. Dr. Shirley', names_dict, tmp_set)
    elif 'Fanny' in name:
        add_alias(name, 'Fanny Harville', names_dict, tmp_set)
    elif 'Trent' in name:
        add_alias(name, 'Governor Trent', names_dict, tmp_set)
    elif 'Alicia' in name:
        add_alias(name, 'Lady Alicia', names_dict, tmp_set)
    elif 'Ives' in name:
        add_alias(name, 'Lord St. Ives', names_dict, tmp_set)
    elif 'Mackenzie' in name:
        add_alias(name, 'Mackenzie', names_dict, tmp_set)
    elif 'Mary' in name:
        add_alias(name, 'Lady Mary Grierson', names_dict, tmp_set)
    elif 'Atkinson' in name:
        add_alias(name, 'Miss Atkinson', names_dict, tmp_set)
    elif 'Carteret' in name:
        add_alias(name, 'Miss Carteret', names_dict, tmp_set)
    elif 'Hamilton' in name:
        add_alias(name, 'Hamilton', names_dict, tmp_set)
    elif 'Robinson' in name:
        add_alias(name, 'Mr. Robinson', names_dict, tmp_set)
    elif 'Shepherd' in name:
        add_alias(name, 'John Shepherd', names_dict, tmp_set)
    elif 'Rooke' in name:
        add_alias(name, 'Nurse Rooke', names_dict, tmp_set)
    elif 'Mrs. Wallis' in name:
        add_alias(name, 'Mrs. Wallis', names_dict, tmp_set)
    elif 'Richard' in name:
        add_alias(name, 'Richard Musgrove', names_dict, tmp_set)
    elif 'Russell' in name:
        add_alias(name, 'Sir Henry Russell', names_dict, tmp_set)
    elif 'Walter' in name:
        add_alias(name, 'Sir Walter Elliot', names_dict, tmp_set)
    elif 'Sophia' in name:
        add_alias(name, 'Sophia Wentworth', names_dict, tmp_set)
    elif 'Lady Russell' in name:
        add_alias(name, 'Lady Russell', names_dict, tmp_set)
 
    
    # ambiguous names
    elif 'Miss Elliot' in name:
        add_alias(name, 'Ambiguous Miss Elliot', names_dict, tmp_set)
    elif 'Miss Harville' in name:
        add_alias(name, 'Ambiguous Miss Harville', names_dict, tmp_set)
    elif 'he' in name:
        add_alias(name, 'Ambiguous Male', names_dict, tmp_set)
    elif 'she' in name:
        add_alias(name, 'Ambiguous Female', names_dict, tmp_set)

# names_dict.keys()
tmp_set
```
