```python
# print(names_set)
names_dict = dict()
tmp_set = names_set.copy()

def add_alias(alias, name, names_dict, tmp_set):
    if name not in names_dict:
        names_dict[name] = list()
    names_dict[name].append(alias)
    tmp_set.remove(alias)
# simple mentions
lst = [
    "Fanny Price", 
    "” “ Miss Price",
    "” “ Fanny Price",
    "Sir Thomas Bertram", 
    "Sir Thomas ’",
    "’ ” Sir Thomas",
    "Lady Bertram", 
    "Edmund Bertram", 
    "Mr. Edmund Bertram", 
    "Mr. Edmund",
    "Maria Bertram", 
    "Miss Maria",
    "Julia Bertram", 
    # "” Miss Bertram",
    "Tom Bertram",
    "Mrs. Norris", 
    "Mrs. Norris ’",
    "Mrs. Norris —",
    "Henry Crawford", 
    "Mr. Crawford ’",
    "Mary Crawford", 
    "Admiral Crawford",
    "Mrs. Crawford",
    # "Miss Crawford ’",
    "Mr. Rushworth",
    "Mr. Rushworth ’",
    "Mr. Rushworth. ”",
    "William Price", 
    "Lady Stornaway", 
    # "Mr. Grant", 
    "Dr. Grant",
    "Mrs. Grant", 
    "Mrs. Rushworth", 
    "Mrs. Price",
    "Thornton Lacey",
    "Thornton Lacey. ”",
    "” Thornton Lacey",
    "Tom Oliver",
    # "Sir Charles",
    "Miss Maddoxes",
    "Mr. Charles Maddox",
    "Miss Owens",
    "Mr. Owen",
    "“ Captain Marshall",
    "Mr. Harding",
    "Mr. Repton",
    "Charles Anderson",
    # "Hon. John Yates",
    "Mr. Yates",
    "Flora Ross",
    "Baron Wildenheim ’",
    




]

for n in lst:
    formatted_n = n.replace("’", "").replace("“","").replace("”", "").replace("—", "").strip()
    add_alias(n, formatted_n, names_dict, tmp_set)
    

complex_set = tmp_set.copy()
for name in complex_set:
        
    if 'Fanny' in name:
        add_alias(name, 'Fanny Price', names_dict, tmp_set)
    elif 'Thomas' in name:
        add_alias(name, 'Sir Thomas Bertram', names_dict, tmp_set)
    elif 'Edmund' in name:
        add_alias(name, 'Mr. Edmund Bertram', names_dict, tmp_set)
    elif 'Thornton' in name:
        add_alias(name, 'Thornton Lacey', names_dict, tmp_set)
    elif 'Maria' in name:
        add_alias(name, 'Maria Bertram', names_dict, tmp_set)
    elif 'Charles' in name:
        add_alias(name, 'Mr. Charles Maddox', names_dict, tmp_set)
    elif 'Maddox' in name:
        add_alias(name, 'Mr. Charles Maddox', names_dict, tmp_set)
    elif 'Thomas' in name:
        add_alias(name, 'Sir Thomas Bertram', names_dict, tmp_set)
    elif 'Norris' in name:
        add_alias(name, 'Mrs. Norris', names_dict, tmp_set)
    elif 'Henry' in name:
        add_alias(name, 'Henry Crawford', names_dict, tmp_set)
    elif 'Yates' in name:
        add_alias(name, 'Hon. John Yates', names_dict, tmp_set)
    # elif '' in name:
    #     add_alias(name, '', names_dict, tmp_set)
    # elif '' in name:
    #     add_alias(name, '', names_dict, tmp_set)
    
    # ambiguous names
    elif 'Miss Bertram' in name:
        add_alias(name, 'Ambiguous Miss Bertram', names_dict, tmp_set)
    elif 'Tom' in name:
        add_alias(name, 'Ambiguous Tom', names_dict, tmp_set)
    elif 'Bertram' in name:
        add_alias(name, 'Ambiguous Bertram', names_dict, tmp_set)


# names_dict.keys()
tmp_set
```
