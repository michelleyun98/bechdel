```python
# print(names_set)
names_dict = dict()
tmp_set = names_set.copy()

def add_alias(alias, name, names_dict, tmp_set):
    if name not in names_dict:
        names_dict[name] = list()
    names_dict[name].append(alias)
    tmp_set.remove(alias)
# simple mentions
lst = ['Captain Carter',
      'Colonel Millar ’',
      'Mr. Hurst',
     'Mr. Hurst.',
     'Mr. Jones',
     'Mr. Jones ’',
     'Mr. Morris',
     'Mr. Philips',
     'Mr. Philips ’',
     'Mr. Robinson',
     'Mr. Robinson ’',
     'Mr. Robinson. ” “',
      'Miss Pope',
     'Miss Watson ’',
     'Miss Webbs',
      'Mrs. Hill',
     'Mrs. Hurst',
     'Mrs. Hurst ’',
     'Mrs. Jenkinson',
     'Mrs. Jenkinson ’',
     'Mrs. Long',
     'Mrs. Nicholls',
     'Mrs. Philips',
     'Mrs. Philips ’',
     'Mrs. Reynolds',
     '” Mrs. Hill',
     '” Mrs. Hurst',
     '” Mrs. Philips',
     '” Mrs. Reynolds ’',
      '” Mr. Gardiner',
      'Mr. Denny',
     'Mr. Gardiner',
     'Mr. Gardiner ’',
     'Mr. Stone',
      'William Goulding',
      'Miss Grantley ’',
      'Sir Lewis',
      'John',
      'Dawson',
      'Lady Metcalfe ’',
      'Richard']

for n in lst:
    formatted_n = n.replace("’", "").replace("“","").replace("”", "").strip()
    add_alias(n, formatted_n, names_dict, tmp_set)
    

complex_set = tmp_set.copy()
for name in complex_set:
        
    
    if 'Lizzy' in name or 'Elizabeth' in name or 'Eliza' in name:
        add_alias(name, 'Eizabeth Bennet', names_dict, tmp_set)
    elif 'Jane' in name:
        add_alias(name, 'Jane Bennet', names_dict, tmp_set)
    elif 'Kitty' in name or 'Catherine Bennet' in name:
        add_alias(name, 'Catherine Bennet', names_dict, tmp_set)
    elif 'Lydia' in name:
        add_alias(name, 'Lydia Bennet', names_dict, tmp_set)
    elif 'Mr. Bennet' in name:
        add_alias(name, 'Mr. Bennet', names_dict, tmp_set)
    elif 'Mrs. Bennet' in name:
        add_alias(name, 'Mrs. Bennet', names_dict, tmp_set)
    elif 'Miss Darcy' in name or 'Georgiana' in name:
        add_alias(name, 'Georgiana Darcy', names_dict, tmp_set)
    elif 'Anne Darcy' in name or 'Lady Anne' in name:
        add_alias(name, 'Anne Darcy', names_dict, tmp_set)
    elif 'Darcy' in name:
        add_alias(name, 'Fitzwilliam Darcy', names_dict, tmp_set)
    elif 'Wickham' in name or 'George' in name:
        add_alias(name, 'George Wickham', names_dict, tmp_set)
    elif 'Charlotte' in name or 'Mrs. Collins' in name or 'Miss Lucas' in name:
        add_alias(name, 'Charlotte Lucas', names_dict, tmp_set)
    elif 'Lady Lucas' in name:
        add_alias(name, 'Lady Lucas', names_dict, tmp_set)
    elif 'Maria' in name:
        add_alias(name, 'Maria Lucas', names_dict, tmp_set)
    elif 'William Lucas' in name or 'Sir William' in name:
        add_alias(name, 'William Lucas', names_dict, tmp_set)
    elif 'Collins' in name:
        add_alias(name, 'Mr. Collins', names_dict, tmp_set)
    elif 'Caroline' in name or 'Miss Bingley' in name:
        add_alias(name, 'Caroline Bingley', names_dict, tmp_set)
    elif 'Louisa' in name or 'Mrs. Hurst' in name:
        add_alias(name, 'Louisa Hurst', names_dict, tmp_set)
    elif 'Bingley' in name or 'Charles' in name:
        add_alias(name, 'Charles Bingley', names_dict, tmp_set)    
    elif 'Anne' in name:
        add_alias(name, 'Anne de Bourgh', names_dict, tmp_set)
    elif 'Bourgh' in name or 'Catherine' in name or 'Ladyship' in name:
        add_alias(name, 'Catherine de Bourgh', names_dict, tmp_set)
    elif 'Colonel Fitzwilliam' in name:
        add_alias(name, 'Colonel Fitzwilliam', names_dict, tmp_set)
    elif 'Colonel Forster' in name or 'Colonel F.' in name:
        add_alias(name, 'Colonel Forster', names_dict, tmp_set)
    elif 'Harriet' in name or 'Mrs. Gardiner' in name:
        add_alias(name, 'Harriet Gardiner', names_dict, tmp_set)
    elif 'Mary' in name or 'King' in name:
        add_alias(name, 'Mary King', names_dict, tmp_set)
    elif 'Mrs. F' in name:
        add_alias(name, 'Mrs. Forster', names_dict, tmp_set)
    elif 'Carter' in name:
        add_alias(name, 'Captain Carter', names_dict, tmp_set)
        
    # ambiguous names
    elif 'Fitzwilliam' in name:
        add_alias(name, 'Ambiguous Fitzwilliam', names_dict, tmp_set)
    elif 'Miss Bennet' in name:
        add_alias(name, 'Ambiguous Miss Bennet', names_dict, tmp_set)
    elif 'Lucas' in name:
        add_alias(name, 'Ambiguous Lucas', names_dict, tmp_set)
    elif 'Gardiner' in name:
        add_alias(name, 'Ambiguous Gardiner', names_dict, tmp_set)
    elif 'Bennet' in name:
        add_alias(name, 'Ambiguous Bennet', names_dict, tmp_set)
    elif 'Philips' in name:
        add_alias(name, 'Ambiguous Philips', names_dict, tmp_set)
# names_dict.keys()
tmp_set
```

```python
female_names = ['Miss Pope', 'Miss Watson', 'Miss Webbs', 'Mrs. Hill', 'Mrs. Hurst', 'Mrs. Jenkinson', 'Mrs. Long', 'Mrs. Nicholls', 'Mrs. Philips', 'Mrs. Reynolds', 'Miss Grantley', 'Dawson', 'Lady Metcalfe', 'Jane Bennet', 'Lydia Bennet', 'Harriet Gardiner', 'Eizabeth Bennet', 'Mrs. Bennet', 'Caroline Bingley', 'Charlotte Lucas', 'Catherine de Bourgh', 'Mary King', 'Ambiguous Miss Bennet', 'Maria Lucas', 'Georgiana Darcy', 'Anne Darcy', 'Catherine Bennet', 'Ambiguous Fitzwilliam', 'Louisa Hurst', 'Anne de Bourgh', 'Mrs. Forster', 'Lady Lucas']

gender_ambiguous_names = ['Ambiguous Bennet', 'Ambiguous Lucas', 'Ambiguous Philips', 'Ambiguous Gardiner']
```

```python
import json
# names_dict
reverse_lookup_dict = dict()
for key, value in names_dict.items():
    for v in value:
        gender = "Male"
        if key in female_names:
            gender = "Female"
        elif key in gender_ambiguous_names:
            gender = "Unknown"
        reverse_lookup_dict[v] = (key, gender)

with open('characters/pride-and-prejudice-characters.txt', 'w') as f:
     f.write(json.dumps(reverse_lookup_dict))
```
